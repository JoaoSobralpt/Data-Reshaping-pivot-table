#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python

df_dt = employee.pivot_table( index = ['Department','Title'], values = ['Salary'], aggfunc= {'Salary': 'mean'}).reset_index()
df_dt.columns = ['Department','Title','Salary_Mean']
df_dt
```
